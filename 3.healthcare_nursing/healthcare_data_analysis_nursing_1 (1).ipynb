### Data pre-processing & Feature Engineering

```python
categorical_cols=df.select_dtypes('object')
for i in categorical_cols:
    print("\n Unique Values")
    print(df[i].value_counts())
```

```python
df = df.copy()

date_cols = ["Date of Admission", "Discharge Date"]
for c in date_cols:
    df[c] = pd.to_datetime(df[c], errors="coerce")

# =========================
# 1. ì—°ë ¹êµ° ë§Œë“¤ê¸° (ìš”ì²­í•œ ë°©ì‹)
# =========================
bn = [10, 18, 25, 65, 90]
label = ["Teens(10-17)", "young Adults(18-24)", "Adults(25-64)", "seniors(>65)"]
df["AgeGroup"] = pd.cut(df["Age"], bins=bn, labels=label, right=False)

# êµ¬ê°„ ë°–/ê²°ì¸¡(ì˜ˆ: 10 ë¯¸ë§Œ, 90 ì´ìƒ)ì€ ì œì™¸í•˜ê±°ë‚˜ ë³„ë„ ì²˜ë¦¬í•´ì•¼ í•¨
df = df[df["AgeGroup"].notna()].copy()

# =========================
# 2. ì—°ë ¹êµ° ë¶ˆê· í˜• ë³´ì • ê°€ì¤‘ì¹˜ ë§Œë“¤ê¸°
#    - ëª©í‘œ(í‘œì¤€) ë¶„í¬: ëª¨ë“  ì—°ë ¹êµ° ë™ì¼ ë¹„ì¤‘
#    - weight = target_prop / observed_prop
# =========================
age_dist = df["AgeGroup"].value_counts(normalize=True).sort_index()
target = pd.Series(1/len(label), index=age_dist.index)

age_weight_map = (target / age_dist).astype(float)
df["age_weight"] = df["AgeGroup"].map(age_weight_map).astype(float)

# (ì„ íƒ) í‰ê·  1ë¡œ ì •ê·œí™”: í•´ì„ íŽ¸ì˜. ë°˜ë“œì‹œ floatì´ì–´ì•¼ í•¨.
df["age_weight"] = df["age_weight"] / df["age_weight"].mean()

# =========================
# 3. í—¬í¼ í•¨ìˆ˜ë“¤ (ê°€ì¤‘ì¹˜ ì ìš©)
# =========================
def wmean(x, w):
    x = pd.to_numeric(x, errors="coerce")
    w = pd.to_numeric(w, errors="coerce")
    m = x.notna() & w.notna()
    if m.sum() == 0:
        return np.nan
    return np.average(x[m], weights=w[m])

def wprop(df_, col, wcol="age_weight"):
    """ë²”ì£¼ë³„ ê°€ì¤‘ ë¹„ìœ¨"""
    s = df_[col].astype("object")
    w = df_[wcol]
    tmp = pd.DataFrame({"cat": s, "w": w}).dropna()
    tot = tmp["w"].sum()
    return (tmp.groupby("cat")["w"].sum() / tot).sort_values(ascending=False)

def wct(df_, row, col, wcol="age_weight", normalize="index"):
    """ê°€ì¤‘ êµì°¨í‘œ"""
    tmp = df_[[row, col, wcol]].dropna()
    pv = tmp.pivot_table(values=wcol, index=row, columns=col, aggfunc="sum", fill_value=0)
    if normalize == "index":
        return pv.div(pv.sum(axis=1), axis=0)
    if normalize == "columns":
        return pv.div(pv.sum(axis=0), axis=1)
    if normalize == "all":
        return pv / pv.values.sum()
    return pv

def add_los(df_):
    out = df_.copy()
    out["LengthOfStay"] = (out["Discharge Date"] - out["Date of Admission"]).dt.days
    return out

df = add_los(df)

# billing ìŒìˆ˜(ìŠ¤í¬ë¦°ìƒ·ì— min < 0): ë¶„ì„ì—ì„œëŠ” ë³´í†µ ê²°ì¸¡/0ì²˜ë¦¬/ì ˆëŒ€ê°’ ë“± ì •ì±…ì´ í•„ìš”.
# ì—¬ê¸°ì„  "ìŒìˆ˜ëŠ” ê²°ì¸¡"ìœ¼ë¡œ ì²˜ë¦¬
df.loc[df["Billing Amount"] < 0, "Billing Amount"] = np.nan
```

```python
# 1. ì—°ë ¹êµ°(ë³´ì •) ë¹„ìœ¨
result_01 = wprop(df, "AgeGroup")

# 2. ì„±ë³„(ë³´ì •) ë¹„ìœ¨
result_02 = wprop(df, "Gender")

# 3. ìž…ì›ìœ í˜•(ë³´ì •) ë¹„ìœ¨
result_03 = wprop(df, "Admission Type")

# 4. ì§ˆí™˜(ë³´ì •) ìƒìœ„ 10 ë¹„ìœ¨
result_04 = wprop(df, "Medical Condition").head(10)

# 5. ê²€ì‚¬ê²°ê³¼(ë³´ì •) ë¹„ìœ¨
result_05 = wprop(df, "Test Results")

# 6. ì—°ë ¹êµ° x ê²€ì‚¬ê²°ê³¼ (ì—°ë ¹ ë¶ˆê· í˜• ë³´ì •ëœ ë¹„êµ)
result_06 = wct(df, "AgeGroup", "Test Results", normalize="index")

# 7. ì„±ë³„ x ê²€ì‚¬ê²°ê³¼
result_07 = wct(df, "Gender", "Test Results", normalize="index")

# 8. ìž…ì›ìœ í˜• x ê²€ì‚¬ê²°ê³¼
result_08 = wct(df, "Admission Type", "Test Results", normalize="index")

# 9. ì§ˆí™˜ë³„ (ì—°ë ¹ë³´ì •) í‰ê·  ì˜ë£Œë¹„
result_09 = df.groupby("Medical Condition").apply(
    lambda g: wmean(g["Billing Amount"], g["age_weight"])
).sort_values(ascending=False)

# 10. ìž…ì›ìœ í˜•ë³„ (ì—°ë ¹ë³´ì •) í‰ê·  ì˜ë£Œë¹„
result_10 = df.groupby("Admission Type").apply(
    lambda g: wmean(g["Billing Amount"], g["age_weight"])
).sort_values(ascending=False)

# 11. ì„±ë³„ë³„ (ì—°ë ¹ë³´ì •) í‰ê·  ì˜ë£Œë¹„
result_11 = df.groupby("Gender").apply(
    lambda g: wmean(g["Billing Amount"], g["age_weight"])
)

# 12. ì§ˆí™˜ë³„ (ì—°ë ¹ë³´ì •) í‰ê·  ìž¬ì›ì¼ìˆ˜(LOS)
result_12 = df.groupby("Medical Condition").apply(
    lambda g: wmean(g["LengthOfStay"], g["age_weight"])
).sort_values(ascending=False)

# 13. ìž…ì›ìœ í˜•ë³„ (ì—°ë ¹ë³´ì •) í‰ê·  ìž¬ì›ì¼ìˆ˜
result_13 = df.groupby("Admission Type").apply(
    lambda g: wmean(g["LengthOfStay"], g["age_weight"])
).sort_values(ascending=False)

# 14. ê²€ì‚¬ê²°ê³¼ë³„ (ì—°ë ¹ë³´ì •) í‰ê·  ìž¬ì›ì¼ìˆ˜
result_14 = df.groupby("Test Results").apply(
    lambda g: wmean(g["LengthOfStay"], g["age_weight"])
).sort_values(ascending=False)

# 15. ë³´í—˜ì‚¬(ë³´ì •) ë¹„ìœ¨ ìƒìœ„ 10
result_15 = wprop(df, "Insurance Provider").head(10)

# 16. ë³´í—˜ì‚¬ x ê²€ì‚¬ê²°ê³¼ (ë³´ì • ë¹„êµ)
result_16 = wct(df, "Insurance Provider", "Test Results", normalize="index")

# 17. í˜ˆì•¡í˜•(ë³´ì •) ë¹„ìœ¨
result_17 = wprop(df, "Blood Type")

# 18. í˜ˆì•¡í˜• x ê²€ì‚¬ê²°ê³¼
result_18 = wct(df, "Blood Type", "Test Results", normalize="index")

# 19. ì•½ë¬¼(ë³´ì •) ë¹„ìœ¨
result_19 = wprop(df, "Medication")

# 20. ì•½ë¬¼ x ê²€ì‚¬ê²°ê³¼
result_20 = wct(df, "Medication", "Test Results", normalize="index")
```

```python
results = {
    "Demographics": {
        "AgeGroup_distribution_weighted": result_01,
        "Gender_distribution_weighted": result_02,
    },
    "Admission": {
        "AdmissionType_distribution_weighted": result_03,
        "AdmissionType_x_TestResults": result_08,
    },
    "Clinical": {
        "Top_MedicalCondition_weighted": result_04,
        "MedicalCondition_x_TestResults": result_06,
        "Gender_x_TestResults": result_07,
    },
    "Cost": {
        "MeanBilling_by_Condition_weighted": result_09,
        "MeanBilling_by_AdmissionType_weighted": result_10,
        "MeanBilling_by_Gender_weighted": result_11,
    },
    "LOS": {
        "MeanLOS_by_Condition_weighted": result_12,
        "MeanLOS_by_AdmissionType_weighted": result_13,
        "MeanLOS_by_TestResults_weighted": result_14,
    },
    "Insurance_Hospital": {
        "Insurance_distribution_weighted": result_15,
        "Insurance_x_TestResults": result_16,
    },
    "Medication_BloodType": {
        "Medication_distribution_weighted": result_19,
        "Medication_x_TestResults": result_20,
        "BloodType_distribution_weighted": result_17,
        "BloodType_x_TestResults": result_18,
    }
}
```
