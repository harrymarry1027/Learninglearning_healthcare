```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
import re
from scispacy.linking import EntityLinker
from tqdm import tqdm

# ==========================================
# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
# ==========================================
print("üìÇ Îç∞Ïù¥ÌÑ∞ Î°úÎî© Ï§ë...")
file_path = '/content/chartevents.csv.gz'  # ÌååÏùº Í≤ΩÎ°ú ÌôïÏù∏

try:
    df = pd.read_csv(file_path, compression='gzip', low_memory=False)
    print(f"‚úÖ Î°úÎìú ÏôÑÎ£å! Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞: {len(df):,}Ìñâ")
except FileNotFoundError:
    print("‚ùå ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§. ÌååÏùºÏù¥ ÏóÖÎ°úÎìúÎêòÏóàÎäîÏßÄ ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî.")
    raise

# ==========================================
# 2. STEP 1: Îç∞Ïù¥ÌÑ∞ Îã§Ïù¥Ïñ¥Ìä∏ (Filtering)
# ==========================================
target_itemid = 220001 # Problem List

# Ìï¥Îãπ itemidÍ∞Ä ÏóÜÏúºÎ©¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Í∞ÄÏû• ÎßéÏùÄ ÌÖçÏä§Ìä∏ Ìï≠Î™©ÏúºÎ°ú ÎåÄÏ≤¥ (ÏóêÎü¨ Î∞©ÏßÄÏö©)
if target_itemid not in df['itemid'].unique():
    print(f"‚ö†Ô∏è ItemID {target_itemid}Í∞Ä ÏÉòÌîåÏóê ÏóÜÏñ¥, Îç∞Ïù¥ÌÑ∞Í∞Ä Í∞ÄÏû• ÎßéÏùÄ ÌÖçÏä§Ìä∏ Ìï≠Î™©ÏúºÎ°ú ÎåÄÏ≤¥Ìï©ÎãàÎã§.")
    text_rows = df[df['valuenum'].isna() & df['value'].notna()]
    target_itemid = text_rows['itemid'].value_counts().idxmax()

print(f"üëâ Î∂ÑÏÑù ÎåÄÏÉÅ ItemID: {target_itemid}")

# ÌïÑÌÑ∞ÎßÅ: ÌäπÏ†ï itemid + ÏàòÏπò Îç∞Ïù¥ÌÑ∞(valuenum) Ï†úÏô∏
df_text = df[(df['itemid'] == target_itemid) & df['valuenum'].isna() & df['value'].notna()].copy()
print(f"-> ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú: {len(df_text):,}Ìñâ")

# ==========================================
# 3. STEP 2: Ï†ÑÏ≤òÎ¶¨ Î∞è ÏãúÍ∞ÅÌôî (Before vs After)
# ==========================================
# (1) Ï†ÑÏ≤òÎ¶¨ Ìï®Ïàò
def clean_text(text):
    text = str(text).lower().strip()      # ÏÜåÎ¨∏Ïûê + ÏïûÎí§ Í≥µÎ∞± Ï†úÍ±∞
    text = re.sub(r'[^a-z0-9\s]', '', text) # ÌäπÏàòÎ¨∏Ïûê Ï†úÍ±∞ (ÏÑ†ÌÉùÏÇ¨Ìï≠)
    text = re.sub(r'\s+', ' ', text)      # Îã§Ï§ë Í≥µÎ∞± Ï†úÍ±∞
    return text

# (2) Ï†ÅÏö©
df_text['value_clean'] = df_text['value'].apply(clean_text)

# (3) ÌÜµÍ≥Ñ ÏÇ∞Ï∂ú
raw_cnt = df_text['value'].nunique()
clean_cnt = df_text['value_clean'].nunique()
reduction = raw_cnt - clean_cnt

# (4) ÏãúÍ∞ÅÌôî: Ïôú Ï†ÑÏ≤òÎ¶¨Î•º ÌñàÎäîÍ∞Ä?
plt.figure(figsize=(12, 5))

# Í∑∏ÎûòÌîÑ A: ÏïïÏ∂ï Ìö®Í≥º
plt.subplot(1, 2, 1)
sns.barplot(x=['Raw (Before)', 'Cleaned (After)'], y=[raw_cnt, clean_cnt], palette=['gray', 'green'])
plt.title(f'Vocabulary Reduction\n({reduction} duplicates merged)', fontsize=14)
plt.ylabel('Unique Phrases Count')
for i, v in enumerate([raw_cnt, clean_cnt]):
    plt.text(i, v, str(v), ha='center', va='bottom', fontweight='bold')

# Í∑∏ÎûòÌîÑ B: ÏÉÅÏúÑ ÌÇ§ÏõåÎìú Î∂ÑÌè¨
plt.subplot(1, 2, 2)
top_words = df_text['value_clean'].value_counts().head(10)
sns.barplot(y=top_words.index, x=top_words.values, palette='viridis')
plt.title('Top 10 Standardized Phrases', fontsize=14)

plt.tight_layout()
plt.show()

print("\nüîé [Î≥ÄÍ≤Ω ÏÇ¨Î°Ä ÌôïÏù∏] Ï†ÑÏ≤òÎ¶¨Î°ú ÌÜµÌï©Îêú ÏºÄÏù¥Ïä§:")
changes = df_text[df_text['value'] != df_text['value_clean']]
if not changes.empty:
    print(changes[['value', 'value_clean']].drop_duplicates().head(5))
else:
    print("Îç∞Ïù¥ÌÑ∞Í∞Ä Ïù¥ÎØ∏ Íπ®ÎÅóÌï©ÎãàÎã§.")

# ==========================================
# 4. STEP 3 & 4: ÎîïÏÖîÎÑàÎ¶¨ Íµ¨Ï∂ï Î∞è CUI Îß§Ìïë
# ==========================================
print("\nüß† Î™®Îç∏ Î°úÎìú Î∞è CUI Îß§Ìïë Ï§ÄÎπÑ Ï§ë...")

# (1) Î™®Îç∏ Î°úÎìú (Î≤ÑÏ†Ñ Ìò∏ÌôòÏÑ± Ï≤¥ÌÅ¨ Ìè¨Ìï®)
try:
    import en_core_sci_sm
    nlp = en_core_sci_sm.load()
    nlp.add_pipe("scispacy_linker", config={"resolve_abbreviations": True, "linker_name": "umls", "threshold": 0.85})
    print("‚úÖ Î™®Îç∏(en_core_sci_sm) Î°úÎìú ÏÑ±Í≥µ!")
except Exception as e:
    print(f"‚ùå Î™®Îç∏ Î°úÎìú Ïã§Ìå®: {e}")
    print("ÌåÅ: 1Îã®Í≥Ñ ÏÑ§Ïπò ÌõÑ 'ÏÑ∏ÏÖò Îã§Ïãú ÏãúÏûë'ÏùÑ ÌïòÏßÄ ÏïäÏïòÎã§Î©¥ ÏóêÎü¨Í∞Ä Î∞úÏÉùÌï©ÎãàÎã§.")
    raise

# (2) Í≥†Ïú† ÎîïÏÖîÎÑàÎ¶¨ ÏÉùÏÑ± (ÏÜçÎèÑ ÏµúÏ†ÅÌôî)
unique_texts = df_text['value_clean'].unique()
print(f"üöÄ Îß§Ìïë ÏãúÏûë (ÎåÄÏÉÅ: {len(unique_texts)}Í∞ú Í≥†Ïú† Î¨∏Ïû•)...")

# (3) Îß§Ìïë Ìï®Ïàò
def get_cui(text):
    doc = nlp(text)
    if doc.ents and doc.ents[0]._.kb_ents:
        cui = doc.ents[0]._.kb_ents[0][0]  # CUI ÏΩîÎìú
        name = doc.ents[0]._.kb_ents[0][1] # ÌëúÏ§Ä Ïù¥Î¶Ñ
        return cui, name
    return None, None

# (4) Ïã§Ìñâ
results = []
for text in tqdm(unique_texts):
    cui, name = get_cui(text)
    results.append({'value_clean': text, 'CUI': cui, 'Standard_Name': name})

# (5) Í≤∞Í≥º Î≥ëÌï©
vocab_df = pd.DataFrame(results)
df_final = df_text.merge(vocab_df, on='value_clean', how='left')

# ==========================================
# 5. ÏµúÏ¢Ö Í≤∞Í≥º ÌôïÏù∏
# ==========================================
print("\n=== ‚ú® ÏµúÏ¢Ö Í≤∞Í≥º Îç∞Ïù¥ÌÑ∞ (Sample) ===")
display(df_final[['charttime', 'value', 'value_clean', 'CUI', 'Standard_Name']].head(10))

success_rate = df_final['CUI'].notna().mean() * 100
print(f"\nüìä CUI Îß§Ìïë ÏÑ±Í≥µÎ•†: {success_rate:.2f}%")
```

```python
# ==========================================
# [STEP 1] CUI Îß§Ìïë Í≤ÄÏ¶ù ÏΩîÎìú
# ==========================================
import pandas as pd

# 1. CUIÏôÄ ÌëúÏ§ÄÎ™ÖÏúºÎ°ú Í∑∏Î£πÌïëÌïòÏó¨, ÏõêÎ≥∏ ÌÖçÏä§Ìä∏(value_clean)Î•º Î¶¨Ïä§Ìä∏Î°ú Î¨∂Í∏∞
cui_group = df_final.groupby(['CUI', 'Standard_Name'])['value_clean'].unique().reset_index()

# 2. ÌëúÌòÑÏù¥ 2Í∞ú Ïù¥ÏÉÅÏù∏ Í≤ÉÎßå ÌïÑÌÑ∞ÎßÅ (Îã§ÏñëÌïú ÌëúÌòÑÏù¥ Ïûò Î¨∂ÏòÄÎäîÏßÄ ÌôïÏù∏)
cui_group['Expression_Count'] = cui_group['value_clean'].map(len)
multi_expr_cuis = cui_group[cui_group['Expression_Count'] > 1].sort_values(by='Expression_Count', ascending=False)

print(f"üîé 2Í∞ÄÏßÄ Ïù¥ÏÉÅÏùò ÌëúÌòÑÏúºÎ°ú Í∏∞Î°ùÎêú ÏßàÎ≥ë/Ï¶ùÏÉÅ: {len(multi_expr_cuis)}Í∞ú Î∞úÍ≤¨\n")

# 3. Í≤∞Í≥º Ï∂úÎ†• (ÏÉÅÏúÑ 20Í∞ú ÌôïÏù∏)
pd.set_option('display.max_colwidth', None)
display(multi_expr_cuis[['CUI', 'Standard_Name', 'value_clean', 'Expression_Count']].head(20))
```

```python
# ==========================================
# [STEP 2-1] Î∂ÄÏ†ïÏñ¥ ÌÉêÏßÄ ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò
# ==========================================
!pip install negspacy

# ==========================================
# [STEP 2-2] ÌååÏù¥ÌîÑÎùºÏù∏Ïóê NegEx Ï∂îÍ∞Ä Î∞è Î°úÏßÅ ÏàòÏ†ï
# ==========================================
import spacy
from negspacy.negation import Negex
from scispacy.linking import EntityLinker

# 1. Î™®Îç∏ Î°úÎìú (Í∏∞Ï°¥ ÏΩîÎìúÏôÄ ÎèôÏùº)
nlp = spacy.load("en_core_sci_sm")

# 2. Negex(Î∂ÄÏ†ïÏñ¥ ÌÉêÏßÄ) Ï∂îÍ∞Ä *** (LinkerÎ≥¥Îã§ Î®ºÏ†Ä Ï∂îÍ∞ÄÌï¥Ïïº Ìï®)
# "ENTITY" ÎùºÎ≤®Ïù¥ Î∂ôÏùÄ Îã®Ïñ¥Ïóê ÎåÄÌï¥ Î∂ÄÏ†ïÏùÑ Í≤ÄÏÇ¨Ìï©ÎãàÎã§.
nlp.add_pipe("negex", config={"ent_types": ["ENTITY"]})

# 3. Entity Linker Ï∂îÍ∞Ä (Í∏∞Ï°¥ ÏΩîÎìúÏôÄ ÎèôÏùº)
nlp.add_pipe("scispacy_linker", config={"resolve_abbreviations": True, "linker_name": "umls"})

print("‚úÖ ÌååÏù¥ÌîÑÎùºÏù∏ Íµ¨ÏÑ± ÏôÑÎ£å: Tokenizer -> NER -> Negex(Î∂ÄÏ†ïÌÉêÏßÄ) -> Linker")

# 4. Îç∞Ïù¥ÌÑ∞ Ï≤òÎ¶¨ Ìï®Ïàò ÏóÖÍ∑∏Î†àÏù¥Îìú (Î∂ÄÏ†ïÏñ¥ ÌïÑÌÑ∞ÎßÅ Í∏∞Îä• Ï∂îÍ∞Ä)
def process_text_with_negation(text):
    doc = nlp(text)
    entities = []

    for ent in doc.ents:
        # *** ÌïµÏã¨ Î≥ÄÍ≤Ω ÏÇ¨Ìï≠: Î∂ÄÏ†ïÏñ¥(negex)Í∞Ä TrueÏù¥Î©¥ ÏàòÏßëÌïòÏßÄ ÏïäÏùå ***
        if ent._.negex:
            continue  # Í±¥ÎÑàÎúÄ (Ïòà: "No pneumonia" -> ÏàòÏßë Ïïà Ìï®)

        if ent._.kb_ents:
            cui = ent._.kb_ents[0][0]  # Í∞ÄÏû• ÎÜíÏùÄ Ï†êÏàòÏùò CUI
            linker = nlp.get_pipe("scispacy_linker")
            standard_name = linker.kb.cui_to_entity[cui].canonical_name

            entities.append({
                "Original_Text": ent.text,
                "CUI": cui,
                "Standard_Name": standard_name
            })
    return entities

# -------------------------------------------------------
# ÌÖåÏä§Ìä∏: Î∂ÄÏ†ïÏñ¥ ÌÉêÏßÄÍ∞Ä Ïûò ÎêòÎäîÏßÄ ÌôïÏù∏
# -------------------------------------------------------
test_text = "Patient has diabetes but no evidence of pneumonia."
print(f"\n[ÌÖåÏä§Ìä∏ Î¨∏Ïû•]: {test_text}")
print(f"[Ï∂îÏ∂ú Í≤∞Í≥º]: {process_text_with_negation(test_text)}")
# ÏòàÏÉÅ Í≤∞Í≥º: diabetesÎäî ÎÇòÏò§Í≥†, pneumoniaÎäî Ïïà ÎÇòÏôÄÏïº ÏÑ±Í≥µ!
```
