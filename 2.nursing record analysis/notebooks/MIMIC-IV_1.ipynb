```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
import re
from scispacy.linking import EntityLinker
from tqdm import tqdm

# ==========================================
# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
# ==========================================
print("üìÇ Îç∞Ïù¥ÌÑ∞ Î°úÎî© Ï§ë...")
file_path = '/content/chartevents.csv.gz'  # ÌååÏùº Í≤ΩÎ°ú ÌôïÏù∏

try:
    df = pd.read_csv(file_path, compression='gzip', low_memory=False)
    print(f"‚úÖ Î°úÎìú ÏôÑÎ£å! Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞: {len(df):,}Ìñâ")
except FileNotFoundError:
    print("‚ùå ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§. ÌååÏùºÏù¥ ÏóÖÎ°úÎìúÎêòÏóàÎäîÏßÄ ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî.")
    raise

# ==========================================
# 2. STEP 1: Îç∞Ïù¥ÌÑ∞ Îã§Ïù¥Ïñ¥Ìä∏ (Filtering)
# ==========================================
target_itemid = 220001 # Problem List

# Ìï¥Îãπ itemidÍ∞Ä ÏóÜÏúºÎ©¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Í∞ÄÏû• ÎßéÏùÄ ÌÖçÏä§Ìä∏ Ìï≠Î™©ÏúºÎ°ú ÎåÄÏ≤¥ (ÏóêÎü¨ Î∞©ÏßÄÏö©)
if target_itemid not in df['itemid'].unique():
    print(f"‚ö†Ô∏è ItemID {target_itemid}Í∞Ä ÏÉòÌîåÏóê ÏóÜÏñ¥, Îç∞Ïù¥ÌÑ∞Í∞Ä Í∞ÄÏû• ÎßéÏùÄ ÌÖçÏä§Ìä∏ Ìï≠Î™©ÏúºÎ°ú ÎåÄÏ≤¥Ìï©ÎãàÎã§.")
    text_rows = df[df['valuenum'].isna() & df['value'].notna()]
    target_itemid = text_rows['itemid'].value_counts().idxmax()

print(f"üëâ Î∂ÑÏÑù ÎåÄÏÉÅ ItemID: {target_itemid}")

# ÌïÑÌÑ∞ÎßÅ: ÌäπÏ†ï itemid + ÏàòÏπò Îç∞Ïù¥ÌÑ∞(valuenum) Ï†úÏô∏
df_text = df[(df['itemid'] == target_itemid) & df['valuenum'].isna() & df['value'].notna()].copy()
print(f"-> ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú: {len(df_text):,}Ìñâ")

# ==========================================
# 3. STEP 2: Ï†ÑÏ≤òÎ¶¨ Î∞è ÏãúÍ∞ÅÌôî (Before vs After)
# ==========================================
# (1) Ï†ÑÏ≤òÎ¶¨ Ìï®Ïàò
def clean_text(text):
    text = str(text).lower().strip()      # ÏÜåÎ¨∏Ïûê + ÏïûÎí§ Í≥µÎ∞± Ï†úÍ±∞
    text = re.sub(r'[^a-z0-9\s]', '', text) # ÌäπÏàòÎ¨∏Ïûê Ï†úÍ±∞ (ÏÑ†ÌÉùÏÇ¨Ìï≠)
    text = re.sub(r'\s+', ' ', text)      # Îã§Ï§ë Í≥µÎ∞± Ï†úÍ±∞
    return text

# (2) Ï†ÅÏö©
df_text['value_clean'] = df_text['value'].apply(clean_text)

# (3) ÌÜµÍ≥Ñ ÏÇ∞Ï∂ú
raw_cnt = df_text['value'].nunique()
clean_cnt = df_text['value_clean'].nunique()
reduction = raw_cnt - clean_cnt

# (4) ÏãúÍ∞ÅÌôî: Ïôú Ï†ÑÏ≤òÎ¶¨Î•º ÌñàÎäîÍ∞Ä?
plt.figure(figsize=(12, 5))

# Í∑∏ÎûòÌîÑ A: ÏïïÏ∂ï Ìö®Í≥º
plt.subplot(1, 2, 1)
sns.barplot(x=['Raw (Before)', 'Cleaned (After)'], y=[raw_cnt, clean_cnt], palette=['gray', 'green'])
plt.title(f'Vocabulary Reduction\n({reduction} duplicates merged)', fontsize=14)
plt.ylabel('Unique Phrases Count')
for i, v in enumerate([raw_cnt, clean_cnt]):
    plt.text(i, v, str(v), ha='center', va='bottom', fontweight='bold')

# Í∑∏ÎûòÌîÑ B: ÏÉÅÏúÑ ÌÇ§ÏõåÎìú Î∂ÑÌè¨
plt.subplot(1, 2, 2)
top_words = df_text['value_clean'].value_counts().head(10)
sns.barplot(y=top_words.index, x=top_words.values, palette='viridis')
plt.title('Top 10 Standardized Phrases', fontsize=14)

plt.tight_layout()
plt.show()

print("\nüîé [Î≥ÄÍ≤Ω ÏÇ¨Î°Ä ÌôïÏù∏] Ï†ÑÏ≤òÎ¶¨Î°ú ÌÜµÌï©Îêú ÏºÄÏù¥Ïä§:")
changes = df_text[df_text['value'] != df_text['value_clean']]
if not changes.empty:
    print(changes[['value', 'value_clean']].drop_duplicates().head(5))
else:
    print("Îç∞Ïù¥ÌÑ∞Í∞Ä Ïù¥ÎØ∏ Íπ®ÎÅóÌï©ÎãàÎã§.")

# ==========================================
# 4. STEP 3 & 4: ÎîïÏÖîÎÑàÎ¶¨ Íµ¨Ï∂ï Î∞è CUI Îß§Ìïë
# ==========================================
print("\nüß† Î™®Îç∏ Î°úÎìú Î∞è CUI Îß§Ìïë Ï§ÄÎπÑ Ï§ë...")

# (1) Î™®Îç∏ Î°úÎìú (Î≤ÑÏ†Ñ Ìò∏ÌôòÏÑ± Ï≤¥ÌÅ¨ Ìè¨Ìï®)
try:
    import en_core_sci_sm
    nlp = en_core_sci_sm.load()
    nlp.add_pipe("scispacy_linker", config={"resolve_abbreviations": True, "linker_name": "umls", "threshold": 0.85})
    print("‚úÖ Î™®Îç∏(en_core_sci_sm) Î°úÎìú ÏÑ±Í≥µ!")
except Exception as e:
    print(f"‚ùå Î™®Îç∏ Î°úÎìú Ïã§Ìå®: {e}")
    print("ÌåÅ: 1Îã®Í≥Ñ ÏÑ§Ïπò ÌõÑ 'ÏÑ∏ÏÖò Îã§Ïãú ÏãúÏûë'ÏùÑ ÌïòÏßÄ ÏïäÏïòÎã§Î©¥ ÏóêÎü¨Í∞Ä Î∞úÏÉùÌï©ÎãàÎã§.")
    raise

# (2) Í≥†Ïú† ÎîïÏÖîÎÑàÎ¶¨ ÏÉùÏÑ± (ÏÜçÎèÑ ÏµúÏ†ÅÌôî)
unique_texts = df_text['value_clean'].unique()
print(f"üöÄ Îß§Ìïë ÏãúÏûë (ÎåÄÏÉÅ: {len(unique_texts)}Í∞ú Í≥†Ïú† Î¨∏Ïû•)...")

# (3) Îß§Ìïë Ìï®Ïàò
def get_cui(text):
    doc = nlp(text)
    if doc.ents and doc.ents[0]._.kb_ents:
        cui = doc.ents[0]._.kb_ents[0][0]  # CUI ÏΩîÎìú
        name = doc.ents[0]._.kb_ents[0][1] # ÌëúÏ§Ä Ïù¥Î¶Ñ
        return cui, name
    return None, None

# (4) Ïã§Ìñâ
results = []
for text in tqdm(unique_texts):
    cui, name = get_cui(text)
    results.append({'value_clean': text, 'CUI': cui, 'Standard_Name': name})

# (5) Í≤∞Í≥º Î≥ëÌï©
vocab_df = pd.DataFrame(results)
df_final = df_text.merge(vocab_df, on='value_clean', how='left')

# ==========================================
# 5. ÏµúÏ¢Ö Í≤∞Í≥º ÌôïÏù∏
# ==========================================
print("\n=== ‚ú® ÏµúÏ¢Ö Í≤∞Í≥º Îç∞Ïù¥ÌÑ∞ (Sample) ===")
display(df_final[['charttime', 'value', 'value_clean', 'CUI', 'Standard_Name']].head(10))

success_rate = df_final['CUI'].notna().mean() * 100
print(f"\nüìä CUI Îß§Ìïë ÏÑ±Í≥µÎ•†: {success_rate:.2f}%")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# ==========================================
# 1. Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ: ÌôòÏûê(ÏûÖÏõê)Î≥ÑÎ°ú Ï¶ùÏÉÅ Î™®ÏúºÍ∏∞
# ==========================================
print("üß© ÌôòÏûêÎ≥Ñ Ï¶ùÏÉÅ Îç∞Ïù¥ÌÑ∞ Î™®ÏúºÎäî Ï§ë")

# CUIÍ∞Ä ÏûàÎäî Îç∞Ïù¥ÌÑ∞Îßå ÏÇ¨Ïö©
df_topic = df_final.dropna(subset=['CUI']).copy()

# (Ï§ëÏöî) 'Acute Pain' Í∞ôÏùÄ Îëê Îã®Ïñ¥ Ïù¥ÏÉÅÏùò Í∞úÎÖêÏù¥ Ï™ºÍ∞úÏßÄÏßÄ ÏïäÎèÑÎ°ù,
# ÌÖçÏä§Ìä∏ ÎåÄÏã† 'CUI ÏΩîÎìú(C00...)'Î•º ÌïôÏäµÏãúÌÇ§Í≥† ÎÇòÏ§ëÏóê Ïù¥Î¶ÑÏúºÎ°ú Î∞îÍøâÎãàÎã§.
# 1) CUI -> Ïù¥Î¶Ñ Îß§Ìïë ÏÇ¨Ï†Ñ ÎßåÎì§Í∏∞ (ÎÇòÏ§ëÏóê Ìï¥ÏÑùÏö©)
cui_to_name = df_topic.set_index('CUI')['Standard_Name'].to_dict()

# 2) ÏûÖÏõê ID(hadm_id)Î≥ÑÎ°ú CUI ÏΩîÎìúÎì§ÏùÑ Ìïú Ï§ÑÎ°ú Ìï©ÏπòÍ∏∞
# Ïòà: "C001 C002 C003"
# (hadm_idÍ∞Ä ÏóÜÏúºÎ©¥ subject_id ÏÇ¨Ïö©)
group_col = 'hadm_id' if 'hadm_id' in df_topic.columns else 'subject_id'
docs = df_topic.groupby(group_col)['CUI'].apply(lambda x: ' '.join(x)).reset_index()

print(f"üëâ Ï¥ù {len(docs)}Í±¥Ïùò ÏûÖÏõê Í∏∞Î°ù(Document) ÏÉùÏÑ± ÏôÑÎ£å")

# ==========================================
# 2. Î≤°ÌÑ∞Ìôî (Îã®Ïñ¥ ÎπàÎèÑ ÌñâÎ†¨ ÎßåÎì§Í∏∞)
# ==========================================
# min_df=2: ÏµúÏÜå 2Î™Ö Ïù¥ÏÉÅÏùò ÌôòÏûêÏóêÍ≤å ÎÇòÌÉÄÎÇú Ï¶ùÏÉÅÎßå Î∂ÑÏÑù (ÎÑàÎ¨¥ Ìù¨Í∑ÄÌïú Í±¥ Ï†úÏô∏)
# (Îç∞Ïù¥ÌÑ∞Í∞Ä ÎÑàÎ¨¥ Ï†ÅÏúºÎ©¥ min_df=1Î°ú ÏûêÎèô Ï°∞Ï†ïÌï©ÎãàÎã§)
min_df_val = 2 if len(docs) > 10 else 1

cv = CountVectorizer(token_pattern=r'(?u)\b\w+\b', min_df=min_df_val)
dtm = cv.fit_transform(docs['CUI'])
feature_names = cv.get_feature_names_out()

print(f"üìê Î∂ÑÏÑù ÌñâÎ†¨ ÌÅ¨Í∏∞: {dtm.shape} (Î¨∏ÏÑú Ïàò x Ï¶ùÏÉÅ Ï¢ÖÎ•ò)")

# ==========================================
# 3. LDA Î™®Îç∏ÎßÅ (Ìå®ÌÑ¥ Ï∞æÍ∏∞)
# ==========================================
# ÌÜ†ÌîΩ Í∞úÏàò(n_components)Îäî 3~5Í∞úÎ°ú ÏÑ§Ï†ïÌï¥ Î¥ÖÎãàÎã§.
n_topics = 3
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(dtm)

# ==========================================
# 4. Í≤∞Í≥º Ìï¥ÏÑù Î∞è ÏãúÍ∞ÅÌôî
# ==========================================
def plot_top_words(model, feature_names, n_top_words=5):
    fig, axes = plt.subplots(1, n_topics, figsize=(15, 6), sharex=True)
    axes = axes.flatten()

    for topic_idx, topic in enumerate(model.components_):
        # Ìï¥Îãπ ÌÜ†ÌîΩÏóêÏÑú Í∞ÄÏû• ÎπÑÏ§ëÏù¥ ÎÜíÏùÄ ÏÉÅÏúÑ CUI Ïù∏Îç±Ïä§ Ï∂îÏ∂ú
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        # CUI ÏΩîÎìúÎ•º Îã§Ïãú Ïö∞Î¶¨Í∞Ä ÏïÑÎäî Ïù¥Î¶Ñ(Standard Name)ÏúºÎ°ú Î≥ÄÌôò
        labels = [cui_to_name.get(cui, cui) for cui in top_features]

        ax = axes[topic_idx]
        ax.barh(labels, weights, height=0.7, color='teal')
        ax.set_title(f'Topic {topic_idx + 1}', fontdict={'fontsize': 16})
        ax.invert_yaxis()
        ax.tick_params(axis='both', which='major', labelsize=10)

        # xÏ∂ï ÎùºÎ≤® Îì± Íæ∏ÎØ∏Í∏∞
        for i in 'top right left'.split():
            ax.spines[i].set_visible(False)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.suptitle(f'Top Symptoms per Topic (CUI Based Analysis)', fontsize=20)
    plt.show()

print("\nüìä ÌÜ†ÌîΩ Î™®Îç∏ÎßÅ Í≤∞Í≥º ÏãúÍ∞ÅÌôî:")
plot_top_words(lda, feature_names)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Îç∞Ïù¥ÌÑ∞ Î°úÎìú (Ï†ÑÏ≤¥ ÌååÏùº Í≤ΩÎ°úÎ°ú ÏàòÏ†ïÌïòÏÑ∏Ïöî)
df = pd.read_csv('/content/chartevents.csv.gz', compression='gzip', low_memory=False)

# 2. ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Îßå ÌïÑÌÑ∞ÎßÅ (ÏàòÏπò Ï†úÏô∏)
df_text = df[df['valuenum'].isna() & df['value'].notna()].copy()
df_text['charttime'] = pd.to_datetime(df_text['charttime'])

# 3. 'Ïã¨Ïû• ÏßàÌôò(Heart)' ÌôòÏûêÍµ∞ ÏãùÎ≥Ñ
# Ïã¨Ïû• Í¥ÄÎ†® ÌÇ§ÏõåÎìúÍ∞Ä Í∏∞Î°ùÎêú ÌôòÏûê(hadm_id)Î•º Ï∞æÏäµÎãàÎã§.
heart_keywords = ['heart failure', 'atrial', 'cardiac', 'myocardial', 'valve', 'angina', 'stenosis', 'rhythm']
pattern = '|'.join(heart_keywords)
heart_patients = df_text[df_text['value'].str.contains(pattern, case=False, na=False)]['hadm_id'].unique()

print(f"Î∞úÍ≤¨Îêú Ïã¨Ïû• ÏßàÌôò Í¥ÄÎ†® ÏûÖÏõê Ïàò: {len(heart_patients):,}Í±¥")

# Ïã¨Ïû• ÌôòÏûê Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂ú
df_heart = df_text[df_text['hadm_id'].isin(heart_patients)].copy()

# 4. ÏûÖÏõê ÏãúÍ∏∞(Early/Middle/Late) Íµ¨Î∂Ñ Î°úÏßÅ
# ÌôòÏûêÎ≥ÑÎ°ú ÏûÖÏõê ÏãúÏûë(min)Í≥º ÎÅù(max) ÏãúÍ∞ÑÏùÑ Íµ¨Ìï¥ÏÑú ÌòÑÏû¨ ÏãúÏ†êÏù¥ Ïñ¥ÎîîÏØ§Ïù∏ÏßÄ Í≥ÑÏÇ∞
time_stats = df_heart.groupby('hadm_id')['charttime'].agg(['min', 'max'])
time_stats['duration'] = time_stats['max'] - time_stats['min']

df_heart = df_heart.merge(time_stats[['min', 'duration']], on='hadm_id')
df_heart['progress'] = (df_heart['charttime'] - df_heart['min']) / df_heart['duration']

def get_stage(progress):
    if progress <= 0.33: return 'Early'   # ÏûÖÏõê Ï¥àÎ∞ò 1/3
    elif progress <= 0.66: return 'Middle' # ÏûÖÏõê Ï§ëÎ∞ò 1/3
    else: return 'Late'                   # ÏûÖÏõê ÌõÑÎ∞ò 1/3

df_heart['stage'] = df_heart['progress'].apply(get_stage)

# 5. ÏãúÍ∏∞Î≥Ñ Ï£ºÏöî Ï¶ùÏÉÅ Ï∂îÏ∂ú Î∞è ÏãúÍ∞ÅÌôî
# Ï†ÑÏ≤òÎ¶¨ (ÏÜåÎ¨∏Ïûê Î≥ÄÌôò)
df_heart['clean_value'] = df_heart['value'].astype(str).str.lower().str.strip()

# ÏãúÍ∏∞Î≥Ñ ÎπàÎèÑÏàò Í≥ÑÏÇ∞
top_symptoms = df_heart.groupby(['stage', 'clean_value']).size().reset_index(name='count')
top_symptoms = top_symptoms.sort_values(['stage', 'count'], ascending=[True, False]).groupby('stage').head(5)

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 6))
sns.barplot(data=top_symptoms, x='stage', y='count', hue='clean_value',
            order=['Early', 'Middle', 'Late'], palette='viridis')
plt.title('Top Symptoms of Heart Patients by Hospital Course')
plt.xlabel('Hospital Stage')
plt.ylabel('Frequency')
plt.legend(title='Symptoms', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import re

# (2) Ï†ÑÏ≤òÎ¶¨ Ìï®Ïàò Ï†ïÏùò
def basic_cleaning(text):
    return str(text).lower().strip()

def advanced_cleaning(text):
    # ÏÜåÎ¨∏Ïûê + Í≥µÎ∞±Ï†úÍ±∞ + ÌäπÏàòÎ¨∏Ïûê Ï†úÍ±∞(ÏïåÌååÎ≤≥, Ïà´Ïûê, Í≥µÎ∞±Îßå ÎÇ®ÍπÄ)
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text) # Îã§Ï§ë Í≥µÎ∞±ÏùÑ ÌïòÎÇòÎ°ú
    return text

# (3) Ï†ÑÏ≤òÎ¶¨ Ï†ÅÏö©
df_text['value_basic'] = df_text['value'].apply(basic_cleaning)
df_text['value_clean'] = df_text['value'].apply(advanced_cleaning)

# (4) Ï†ïÍ∑úÌôî Ìö®Í≥º Î∂ÑÏÑù
clean_unique = df_text['value_clean'].nunique()
reduction = raw_unique - clean_unique
```

```python
# --- ÏãúÍ∞ÅÌôî: Ï†ÑÏ≤òÎ¶¨Î•º Ïôú ÌñàÎäîÍ∞Ä? ---
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Í∑∏ÎûòÌîÑ 1: Îç∞Ïù¥ÌÑ∞ ÏïïÏ∂ï Ìö®Í≥º (Bar Plot)
sns.barplot(x=['Before (Raw)', 'After (Clean)'], y=[raw_unique, clean_unique], ax=axes[0], palette=['gray', 'green'])
axes[0].set_title(f'Vocabulary Reduction (Merged {reduction} duplicates)', fontsize=14)
axes[0].set_ylabel('Number of Unique Phrases')
for i, v in enumerate([raw_unique, clean_unique]):
    axes[0].text(i, v, str(v), ha='center', va='bottom', fontweight='bold', fontsize=12)

# Í∑∏ÎûòÌîÑ 2: Ï†ïÍ∑úÌôîÎ°ú Ïù∏Ìï¥ Ìï©Ï≥êÏßÑ ÏÇ¨Î°Ä Ï∞æÍ∏∞ (Top Changes)
# ÏõêÎ≥∏Í≥º Ï†ïÍ∑úÌôîÎ≥∏Ïù¥ Îã§Î•∏ Í≤ΩÏö∞Îßå Ï∂îÏ∂ú
changes = df_text[df_text['value'] != df_text['value_clean']]
if not changes.empty:
    top_changes = changes['value'].value_counts().head(50)
    sns.barplot(y=top_changes.index, x=top_changes.values, ax=axes[1], palette='Reds_r')
    axes[1].set_title('Top 10 "Messy" Texts Corrected', fontsize=14)
    axes[1].set_xlabel('Count')
else:
    axes[1].text(0.5, 0.5, "Data was already clean!", ha='center')

plt.tight_layout()
plt.show()

# Ïã§Ï†ú Î≥ÄÍ≤Ω ÏÇ¨Î°Ä Ï∂úÎ†•
print("\n=== [Ï†ÑÏ≤òÎ¶¨ ÏÉÅÏÑ∏ ÌôïÏù∏] ÌÖçÏä§Ìä∏Í∞Ä Ïñ¥ÎñªÍ≤å Î≥ÄÌñàÎäîÍ∞Ä? (ÏÉòÌîå 5Í∞ú) ===")
print(df_text[['value', 'value_clean']].drop_duplicates().head(5))
```
